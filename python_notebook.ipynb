3.5 Updateing a Single Value

```python
#if we just want to update one piece of a data in a column we can use iloc
df.iloc[4, -7] = 'No'
#df.iloc [0#row, -7 for last column] = 'value'
df.head()
```

3.6 Condition based Updateing using Apply

```python
#if want to take our churn column and represent it as 1's and 0's
#we will need to filter it first with the apply meathod
df['Churn Binary'] = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
df.head()
```
